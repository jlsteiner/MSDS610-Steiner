One of the biggest issues with a dirty dataset is getting an accurate idea of what the counts are for each category.  For example, with the countries--I only did a base level cleanup on those and I still could have done a lot more.  The large number of nulls also make it difficult to assess the data for value.  Dirty data prevents the ML from being able to learn, as well.  Null isn't a value--it just doesn't exist.  The ML can't do anything with it.  The initial data set appeared to have at least one null in each record.  Moreover, various features contained a mixed bag of values--several different values for USA, including misspellings and added punctuation.  Some of the entries had extra spaces in the free form text entries.  This creates a skewed count and will create an erroneous result from the ML, because it will see each one as a distinct entity.  
<br>
Removing the nulls and removing the columns that do not add value to the dataset, will improve the ML's ability to learn from the data. Without the nulls, the ML will be able to use all the columns to perform training and analytics.  However, given that the dataset shows a bias for whites, women, master's level education or higher, and for those in the US--this indicates that prior to training this dataset, the data person will need to add 'mock' data for men, for other races, other education levels and other nations based on the existing data.  If this is not done, the outcome of the ML will be bias and it will remain that way until the data is balanced.  

Anytime data is touched/updated/cleaned it does skew it.  Had the original dataset been used, much of it would have been useless for ML.  Cleaned, the ML now has more to work with.  The question is whether it skews it for better or worse.  The best thing to do when cleaning data is to make as little impact as possible, by imputing accurate values.  The drawback with this is that it can be excessively time consuming to impute every single questionable dirty value.
