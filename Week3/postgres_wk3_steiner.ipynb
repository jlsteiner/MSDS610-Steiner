okay--so it looks like when there was no data available for a question/category null was entered.   I don't want to remove these, since it could skew the data.  Let's check to see what value could be given to the NaN.  Will it ve feasible to move the NaN values to 0?  I want to check the Foot notes to see if there is any insights there.

```python
clean_d = dis.copy()
```

```python
clean_d.DataValueFootnote.unique()
```

```python
clean_d[clean_d['DataValue'].isnull()]
```

It appears that for those with a NaN data value, it should be safe to set those 0.  The footnotes imply that for one reason or another the data was not available or not usable

```python
clean_d[clean_d['DataValueFootnote'].isnull() ]
```

```python
clean_d['DataValue'] = clean_d['DataValue'].fillna(0)
```

```python
clean_d.head()
```

```python
clean_d.info()
```

```python
clean_d['DataValueFootnote'] = clean_d['DataValueFootnote'].fillna('None')
```

```python
clean_d.DataValueFootnote.unique()
```

The value of None is no longer a null, but a string value.  
this means that as data is processed, records with None will not be elimiated.

```python
table_name = r'\disease_cleaned'
schema = r'cleaned' # schema we created this week.

clean_d.to_sql(table_name, con=db_conn, if_exists='replace', index=False, schema=schema, chunksize=1000, method='multi')
```
